```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
```

```python
Driver = webdriver.Chrome("C:/Users/Diya/Downloads/chromedriver_win32 (1)/chromedriver.exe")
url= 'https://en.wikipedia.org/wiki/Narendra_Modi'
Driver.get(url)
elems = Driver.find_elements(By.TAG_NAME, "a")
links=[]
for i in elems:
    href = i.get_attribute("href") 
    if href is not None:
        links.append(href)
print(links)
```

```python
lnks=[]
for j in links:
    new_driver= webdriver.Chrome("C:/Users/Diya/Downloads/chromedriver_win32 (1)/chromedriver.exe")
    new_driver.get(j)
    print("\n", j, "\n")
    elm = new_driver.find_elements(By.TAG_NAME, "a")
    time.sleep(1)
    for t in elm:
        hrf = t.get_attribute("href") 
        if hrf is not None:
            lnks.append(hrf)
    print(lnks)
    lnks.clear()
    new_driver.close()
```
